```python
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go


x_arr = np.random.randn(100)
y_arr = np.random.randn(100)


fig1 = px.line(x= x_arr, y= y_arr)
fig2 = px.scatter(x= x_arr, y= y_arr)

fig3= go.Figure(data=fig1.data + fig2.data)
fig3.show()
```
