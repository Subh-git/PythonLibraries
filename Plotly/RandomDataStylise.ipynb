```python
import pandas as pd
import plotly.express as px
import numpy as np

x_arr = np.random.randn(500)
y_arr = np.random.randn(500)

df = pd.DataFrame({'X': x_arr.tolist(),
'Y': y_arr.tolist()})

fig = px.scatter(df, x= 'X', y= 'Y')
fig.update_traces(marker=dict(size=7,
                                color= 'LightSkyBlue',
                              line=dict(width=1.5,
                                        color='MediumPurple')),
                  selector=dict(mode='markers'))

```
