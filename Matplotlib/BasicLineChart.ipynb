```python
from matplotlib import pyplot as plt   #This is the import satement that we should be using with matplotlib

ages_x = [20,21,25,26,28,29,30,35,36,42,45,51]
salary_y = [3200,4200,4600,4700,5100,5250,5600,5800,6000,7500,7800,8000]

plt.plot(ages_x, salary_y, color='b', linestyle ='--', linewidth = 3, label = 'Average Salary')  #as the x and y axis
#we could also pass hex color values by searching from the internet.

plt.xlabel('Ages')
plt.ylabel("Average Salary(USD)")
plt.title("Average Salary by Age")

plt.legend()   #shows the legend
plt.show()
```
