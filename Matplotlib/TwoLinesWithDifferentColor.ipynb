```python
from matplotlib import pyplot as plt


year_x = [2010,2012,2014,2016,2018,2020]
pydev_y = [40000,45000,55000,58000,67000,80000]
jsdev_y = [50000,55000,60000,63000,65000,70000]

plt.title("Salary in years")
plt.plot(year_x, pydev_y, color = 'r', linewidth = 3, label = 'Python Dev')
plt.plot(year_x, jsdev_y, color = 'y', linewidth = 3,label = 'JavaScript Dev')
plt.legend()
plt.show()
```
