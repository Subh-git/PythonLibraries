```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_style('whitegrid')
df = pd.read_csv('expectancy.csv')

sns.boxplot(x='continent', y= 'lifeExp',data= df)
plt.title("Life Expectancy  agaisnt Continent")
plt.show()
```
