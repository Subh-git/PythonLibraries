```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_style('whitegrid')
df = sns.load_dataset('tips')

sns.boxplot(x='day', y= 'tip',data= df)
plt.title("Day Vs Tips")
plt.show()
```
