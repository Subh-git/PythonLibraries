```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

df = sns.load_dataset('tips')


sns.swarmplot(x='size', y= 'total_bill', data= df, size = 3.6)
plt.title("Total Bill against the Size")
plt.show()
```
