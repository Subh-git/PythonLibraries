```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style('darkgrid')
titanic_df = sns.load_dataset('titanic')

sns.pointplot(x = 'sex', y = 'survived', data= titanic_df, color= 'brown', ci = 95)
#ci or confidence interval tells us how sure the data that we perform analysis on is accurate. It depends on various factors

plt.title("Sex Against Survived")
plt.show()
```
