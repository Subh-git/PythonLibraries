```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


#scatterplot is know as relational plot. That tells us the positive or negative trend.

tipdf = sns.load_dataset('tips')
sns.set_style('darkgrid')

sns.scatterplot(x = 'day', y= 'total_bill', data= tipdf, color = 'blue')

plt.title("Day Against Total Bill")
plt.show()


```
