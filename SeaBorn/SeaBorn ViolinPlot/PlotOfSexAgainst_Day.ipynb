```python
import seaborn as sns
import matplotlib.pyplot as plt


#violin plot is a categorical distribution plot
#almost similar to box plot

tipdf = sns.load_dataset('tips')
#print(tipdf)

sns.set_style('whitegrid')

sns.violinplot(x='sex', y='total_bill', data= tipdf)
plt.title("Sex against total bill")
plt.show()
```
