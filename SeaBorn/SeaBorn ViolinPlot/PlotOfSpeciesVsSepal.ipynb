```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.violinplot(x= 'species',y= 'sepal_length', data= iris, palette= 'muted')


plt.title("Species Against sepal length")
plt.show()
```
