```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style('darkgrid')
titanic_df = sns.load_dataset('titanic')     #its inbuilt so no need to import
#ci = confidence interval. can be changed while plotting.
#bar plots are categorical distribution, meaning relation to some numerical value
sns.barplot(x='sex', y = 'survived', data = titanic_df, edgecolor = 'black')

plt.title("Sex of People Vs Survived")               #We can use plt titles also along with seaborn as sns is an extn of plt
plt.show()
```
