```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime


data = pd.read_csv('data.csv')
year = data['Time']
sea_levels = data['GMSL']
plt.scatter(year, sea_levels, edgecolors='g')

plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sealevel')
plt.show()
```
