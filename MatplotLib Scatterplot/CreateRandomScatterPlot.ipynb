```python

import matplotlib.pyplot as plt
from pylab import randn
X = randn(200)
Y = randn(200)
plt.scatter(X,Y, color='r')
plt.xlabel("X")
plt.title("Random distribution")
plt.ylabel("Y")
plt.show()
```
