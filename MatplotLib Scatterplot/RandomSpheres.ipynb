```python
import numpy as np
import matplotlib.pyplot as plt
import random

num = 30
X = np.array([])
Y = np.array([])
for i in range(30):
    x_rd = random.random()
    y_rd = random.random()
    Y = np.append(Y,y_rd)
    X = np.append(X,x_rd)

colors = [random.randint(1, 4) for i in range(num)]
size = np.array([])
for i in range(30):
    size = np.append(size,i*2.5)

plt.scatter(X,Y, c = colors, s = size, alpha= 0.85)
plt.xlabel("X")
plt.title("Random distribution")
plt.ylabel("Y")
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.show()
```
